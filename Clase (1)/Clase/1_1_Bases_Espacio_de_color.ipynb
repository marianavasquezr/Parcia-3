```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('cell.jpg')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
print ('tamanho, producto de f*col*can ', img.size)
print ('tipo                           ', img.dtype)
print ('tamanho de la primera dimension', len(img))
print ('tamanho de la segunda          ', len(img[0,:,:]))
print ('tamanho de la tercera          ', len(img[0,0,:]))
print ('filas, columnas y canales      ', img.shape)
print ('media                          ', img.mean())
print ('desviacion estandar            ', img.std())
print ('valor minimo                   ', img.min())
print ('valor maximo                   ', img.max())
```

```python
plt.imshow(img)
plt.show()
# cv2.namedWindow('Ventana', 0)   #Crear ventana
# cv2.imshow('Ventana',img)       #Mostrar la imagen en la ventana creada
# cv2.waitKey(0)                  #Dejar la ventana abierta hasta el siguiente enter
# cv2.destroyAllWindows() 
```

```python
imgR=img[:,:,0]
imgG=img[:,:,1]
imgB=img[:,:,2]
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,3,1)
plt.imshow(imgB, cmap='gray')
plt.subplot(1,3,2)
plt.imshow(imgG, cmap='gray')
plt.subplot(1,3,3)
plt.imshow(imgR, cmap='gray')
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,3,1)
plt.imshow(imgB, cmap='Reds')
plt.subplot(1,3,2)
plt.imshow(imgG, cmap='Greens')
plt.subplot(1,3,3)
plt.imshow(imgR, cmap='Blues')
```

```python
# Elimina canales para graficar el color real de cada canal
imaR=np.copy(img)
imaR[:,:,1]=0
imaR[:,:,2]=0

imaG=np.copy(img)
imaG[:,:,0]=0
imaG[:,:,2]=0

imaB=np.copy(img)
imaB[:,:,1]=0
imaB[:,:,0]=0
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,3,1)
plt.imshow(imaB)
plt.subplot(1,3,2)
plt.imshow(imaG)
plt.subplot(1,3,3)
plt.imshow(imaR)
```
