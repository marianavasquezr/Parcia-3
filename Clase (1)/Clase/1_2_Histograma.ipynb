```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('Can1.png')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
def histograma(var):
    histR=np.zeros(256)
    histG=np.zeros(256)
    histB=np.zeros(256)
    
    for i in range(np.shape(img)[0]):
        for j in range(np.shape(img)[1]):
            histR[var[i,j,0]]=histR[var[i,j,0]]+1    
            histG[var[i,j,1]]=histG[var[i,j,1]]+1 
            histB[var[i,j,2]]=histB[var[i,j,2]]+1 
            
    return histR, histG, histB 
```

```python
print(np.shape(img)[0])
print(np.shape(img)[1])
print(np.size(img[:,:,0]))
print(np.size(img[:,:,1]))
print(np.size(img[:,:,2]))
print(np.size(img))
```

```python
Hr, Hg, Hb=histograma(img)
```

```python
def graficaHis(R, G, B):    
    plt.plot(R, 'r')
    plt.plot(G, 'g')
    plt.plot(B, 'b')
    plt.title('Histograma')
    plt.xlabel('Niveles')
    plt.ylabel('Densidad')
    plt.grid()
    plt.xlim((0, 255))
    plt.show()
```

```python
graficaHis(Hr, Hg, Hb)
```

```python
print('Valores totales ', np.sum(Hb), np.sum(Hg), np.sum(Hr))
print('Pixeles totales ', np.sum(Hb)+np.sum(Hg)+np.sum(Hr), np.size(img))
```
