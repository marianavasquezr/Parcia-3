### Mostrar imagen
Existen diferentes formas de mostrar una imagen, la primera de ellas es utilizando las librer칤as de matplotlib la cual est치 contenida en el espacio de color RGB, y la segunda de ellas es con las librer칤as de Opencv, contenidas en el espacio BGR.

Los conos son los sensores en el ojo responsables de la visi칩n del color. Evidencia experimental detallada ha establecido que los 6 a 7 millones de conos en el ojo humano se pueden dividir en tres categor칤as principales de detecci칩n, que corresponde aproximadamente a rojo (65%), verde (33%), y azul (2%), pero los conos azules son los m치s sensibles.
Los colores pueden considerarse como una combinaci칩n variable de los llamados colores primarios rojo (R), verde (G) y azul (B); por ejemplo magenta (rojo m치s azul), cian (verde, m치s azul), y amarillo (rojo m치s verde). Con el prop칩sito de la normalizaci칩n, la CIE (Comisi칩n Internacional de la Iluminaci칩n) design칩 los siguientes valores de longitudes de
onda espec칤ficas a los tres colores primarios: 洧녩洧녾洧녹洧녳 = 435,8 洧녵洧녴, 洧녺洧뉧롐洧녬洧 = 546,1 洧녵洧녴, y 洧洧녶洧녱洧녶 = 700 洧녵m(Gonz치lez y Woods, 2002). Las caracter칤sticas generalmente utilizadas para distinguir un color de otro son el brillo, el tono y la saturaci칩n. 

El brillo se refiere a la noci칩n acrom치tica de intensidad,

el tono (hue) es un atributo que se refiere a la longitud de onda dominante en una mezcla de ondas de luz (color dominante) y 

la saturaci칩n se refiere a la cantidad de luz blanca mezclada con una tonalidad (pureza relativa). 

El tono y la saturaci칩n representan juntos la cromaticidad, que es un atributo de la sensaci칩n visual, seg칰n el cual, un 치rea observada aparece m치s o menos
coloreada. 

Advertencia: Como la imagen se lee con las librer칤as de opencv estas se leen por defecto en el espacio BGR.

```python
#Mostrar imagen con matplotlib
#cambio de espacio de color
imap=cv2.cvtColor(ima,cv2.COLOR_BGR2RGB)

#grafica por defecto
plt.imshow(imap)
plt.show()
```
